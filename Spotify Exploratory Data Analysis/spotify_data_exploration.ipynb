<h1>Spotify Exploratory Data Anaslysis<h1>

```python
#Importing relevant libraries
import warnings
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
warnings.filterwarnings('ignore')
```

```python
#Loading the data
tracks_data = pd.read_csv('./data/tracks.csv')
tracks_data.head()
```

```python
#Checking for null values
pd.isnull(tracks_data).sum()
```

```python
tracks_data.info()
```

```python
#sorting data by popularity
sorted_data = tracks_data.sort_values('popularity',ascending=True).head(10)
sorted_data
```

```python
tracks_data.describe().transpose()
```

```python
#Most Popular Tracks
most_popular_tracks = tracks_data.query('popularity>90',inplace=False).sort_values('popularity',ascending=True)
most_popular_tracks[:10]
```

```python
tracks_data.set_index("release_date", inplace=True)
```

```python
#Changing 'release_date' column to datetime
tracks_data.index = pd.to_datetime(tracks_data.index)
tracks_data.head()
```

```python
#Creating a new column which shows the duration of songs in seconds
tracks_data['duration'] = tracks_data['duration_ms'].apply(lambda x: round(x/1000))

#Dropping the original 'duration_ms' column
tracks_data.drop('duration_ms', inplace=True, axis=1)
```

```python
tracks_data.duration.head()
```

```python
#Finding the correlation between the features
corr_data = tracks_data.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))

#plotting a heatmap to visualize the correlations
heatmap = sns.heatmap(corr_data,annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="Greens", linewidths=1, linecolor='Black')
heatmap.set_title("Correlation HeatMap Between Variables")

heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_data = tracks_data.sample(int(0.004*len(tracks_data)))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_data, y = 'loudness', x = 'energy', color = 'g').set(title = 'Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_data, y = 'popularity', x = 'acousticness', color = 'g').set(title = 'Popularity vs Accousticness')
```

```python
tracks_data['dates'] = tracks_data.index.get_level_values('release_date')
tracks_data.dates = pd.to_datetime(tracks_data.dates)
years = tracks_data.dates.dt.year
```

```python
sns.displot(years, color='green', discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per year')
```

```python
total_dur = tracks_data.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dur, ax = ax, color='green', errwidth = False).set(title='Year vs Duration')
plt.xticks(rotation=60)
```

```python
#Loading data on the music genres
genre_data = pd.read_csv('./data/SpotifyFeatures.csv')
genre_data.head()
```

```python
plt.title("Duration of the Songs in Different Genres")
sns.barplot(y='genre', x='duration_ms',color="green" ,data= genre_data)
plt.xlabel("Duration in milli seconds")
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous = genre_data.sort_values("popularity",ascending = False).head(10)
sns.barplot(y='genre', x='popularity', color="green", data = famous).set(title='Top 5 Genres by Popularity')
```

```python
#Loading data on the artiste
artiste_data = pd.read_csv('./data/artists.csv')
artiste_data.head()
```

```python
artiste_data.sort_values('popularity',ascending=True)
artiste_data.set_index('name', inplace=True)
```

```python
#Most Popular Artiste
most_popular_artiste = artiste_data.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_popular_artiste[['name','followers','popularity']][:10]
```

```python
#Most Followed Artiste
most_followed_artiste = artiste_data[['name','followers','popularity']].sort_values('followers',ascending=False)[:10]
most_followed_artiste
```

```python
sns.set_style(style="darkgrid")

plt.figure(figsize=(10,5))

sns.barplot(y='name', x='followers', color="green", data = most_followed_artiste).set(title='Top 10 Most Followed Artiste')
```
